# document_TF

```python
#document_TF():
document_TF = []
for i in file_list:
    array = []
    for j in word:
        array.append(i.count(j))
    document_TF.append(array)    
```

# idf

```python
#idf
idf = []

for i in word:
    idfcount = 0
    for j in file_list:
        if i in j:
            idfcount = idfcount + 1    
    idf.append(idfcount)       
```

# tfiq

```python
K3 = 1000

tfiq = []
for i in range(len(query_list)):
    array = []
    for j in range(len(word)):
        B = float(((K3 + 1) * query_TF[i][j]) / (K3 + query_TF[i][j]))
        array.append(B)
    tfiq.append(array)   
```

```python
tfiq[0]
```

# tfij

```python
K1 = 1.5
b = 1

tfij = []
for i in range(len(file_list)):
    array = []
    for j in range(len(word)):
        A = float(((K1+1)*document_TF[i][j]) / (K1*((1-b)+b*len(file_list[i].split())/173.5) + document_TF[i][j]))
        array.append(A)
    tfij.append(array)   
```

```python
tfij[0]
```

# idfWeighting

```python
idfWeighting = []
for i in idf:
    a = math.log((2265 + 0.5 ) / (float(i) + 0.5 ))
    idfWeighting.append(a)
idfWeighting    
```

```python
queryList = []
for i in range(len(query_list)):
    documentList = []
    for j in range(len(file_list)):
        dic = {}
        sim = 0
        for k in range(len(word)):
            sim = sim + float(tfij[j][k])*float(tfiq[i][k])*float(idfWeighting[k])
            
        dic['fileName'] = Doclist[j]
        dic['sim'] = sim
        documentList.append(dic)
    documentList.sort(key=lambda k: k['sim'], reverse=True)
    
    oneLine = {}
    oneLine["queryName"] = lines[i]
    oneLine["content"] = ""
    
    for queryRes in documentList:
        oneLine["content"] = oneLine["content"] + queryRes['fileName'] + " "
        
    queryList.append(oneLine)    
            
            
```

```python
f = open("C:/Users/123/Desktop/資料檢索_重寫/資料檢索作業.txt" , 'w')

f.write("Query,RetrievedDocuments" )
f.write('\n')
for value in queryList:
    f.write(value["queryName"] + "," + value["content"] + "\r\n")
    f.write('\n')

f.close()
```
